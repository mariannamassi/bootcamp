```
def para_dia(mes):
  ano = int(mes[:4])
  if mes =='2008/Fev':
    return date(2008,2,1)
    return date(ano,1,1)
mensal.index.map(para_dia)
```

```
meses = {
    "Jan" : 1,
    "Fev" : 2,
    "Mar" : 3,
    "Abr" : 4,
    "Mai" : 5,
    "Jun" : 6,
    "Jul" : 7,
    "Ago" : 8,
    "Set" : 9,
    "Out" : 10,
    "Nov" : 11,
    "Dez" : 12
}


def para_dia(ano_mes: str):
  ano: int = int(ano_mes[:4])
  mes: str = ano_mes[5:]
  mes_numero: int = meses[mes]
  return date(ano,mes_numero, 1)
mensal.index = mensal.index.map(para_dia)
mensal.head()
```

```
mensal_melted["mes"] = mensal_melted["dia_mes_ano"].dt.month
mensal_melted["ano"] = mensal_melted["dia_mes_ano"].dt.year
mensal_melted.head()
```
