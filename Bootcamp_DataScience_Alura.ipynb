# Aquisição e Leitura dos Dados Reais

##Material de Aula

```
#Importar a biblioteca Pandas
import pandas as pd

#Utilizar o link do dataset bruto (raw)
fonte = "https://raw.githubusercontent.com/alura-cursos/agendamento-hospitalar/main/dados/A151346189_28_143_208.csv"

#Ler os dados em CSV
#**Para tornar os dados br legíveis**
#Enconding = Permitir caractere brasileiro
#Skiprows = Pular as últimas 3 linhas
#Sep = separar por (string)
#Skipfooter = Pular rodapé
#Thousands = separar centenas por . (string)
#Decimal = separar decimal por , (string)

dados = pd.read_csv(fonte, encoding = "ISO-8859-1",
                    skiprows=3, sep=";", skipfooter=12,
                    thousands=".", decimal=",")
dados.head()
```

```
#Média dos dados
dados.mean()
```

```
#Mudar formato do número
pd.options.display.float_format = "{:.2f}".format
```

```
#Média
dados.mean()
```

```
#Plotar com o plt 
import matplotlib.pyplot as plt

dados.plot(x="Unidade da Federação", y="2008/Ago", kind="bar", figsize=(9,6))

plt.title("Valor por UF")
plt.ylabel("Valor")
```

```
#Plotar com o plt 
import matplotlib.pyplot as plt
#Importar o ticker para formatar o valor
import matplotlib.ticker as ticker

ax = dados.plot(x="Unidade da Federação", y="2008/Ago", kind="bar", figsize=(9,6))
#Formatar o valor do eixo Y em milhões
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))

plt.title("Valor por UF")
plt.ylabel("Valor")
```

```
#Plotar com o plt 
import matplotlib.pyplot as plt
#Importar o ticker para formatar o valor
import matplotlib.ticker as ticker

ax = dados.plot(x="Unidade da Federação", y="2008/Ago", kind="bar", figsize=(9,6))
#Formatar o valor do eixo Y em milhões
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))

plt.title("Valor por UF")
plt.ylabel("Valor")

#Para ocultar a última função
plt.show()
```

# Desafios

##Desafio 01: Escolher um título mais descritivo, que passe a mensagem adequada ao gráfico de barras.

```
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

ax = dados.plot(x="Unidade da Federação", y="2008/Ago", kind="bar", figsize=(9,6))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))

#Alterar o título
plt.title("Produção Hospitalar em Agosto de 2008 por Estado Brasileiro")

plt.xlabel("Estado")
plt.ylabel("Produção Hospitalar")

plt.show()
```

##Desafio 02: Faça a mesma análise realizada em aula, porém para o mês mais recente.

```
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

#Alterar para a coluna referente ao mês mais recente (agosto não possui dados suficientes)
ax = dados.plot(x="Unidade da Federação", y="2020/Jul", kind="bar", figsize=(9,6))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))

#Alterar o título para "Julho de 2020"
plt.title("Produção Hospitalar em Julho de 2020 por Estado Brasileiro")

plt.xlabel("Estado")
plt.ylabel("Produção Hospitalar")

plt.show()
```
